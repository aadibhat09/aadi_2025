# NodCursor at the CTE Expo - Reflections & Takeaways
 
## Overview
 
This issue documents my experience presenting NodCursor at the CTE Expo, including what I demoed, conversations I had with attendees, and key takeaways from the event.
 
---
 
## What is NodCursor?
 
The nodcursor project aims at enhancing accessibility in that it incorporates the use of MediaPipe technology in detecting a user's head location and movement in order to navigate their computer cursor. It gives an opportunity for people with motor disabilities to be able to maneuver their computers using only their heads.
 
This was initially developed as an individual website using the MediaPipe face landmarks to track the head's position and convert that into cursor movement. Later on, it was implemented into a platform that is widely used by students in the Computer Science classes at my university, known as the Open Coding Society (OCS) platform. This helped create awareness that implementing accessibility features in a project is possible and feasible.
 
---
 
## What I Demoed
 
At the Expo, I demoed a maze game built on top of NodCursor. Players navigate through the maze entirely using their head.

The maze game served two purposes:
 
1. User experience – It was an engaging experience for those who tried it out as it allowed them to understand what it feels like to do something using head tracking technology. It made the use of head tracking technology more real since you were able to do it yourself.

2. Developer communication – This was my way of opening the discussion on the feasibility of adding head tracking technology into their apps or programs. What if they realized that they too can make something work using head tracking?

Some of the highlights of the experience were the opportunity to showcase alongside Nikhil, Rohan, and Aditya for their creation of a facial recognition technology used to create a pass that allows access to a digitally-equipped bathroom. While we worked on projects that differed in scale and problem area, they did make us see what kind of work can be done by students who took CS classes. Our table had a lot of interest from the attendees not because of social politeness but because of genuine curiosity and desire to try it out ourselves.
 
Having people play with NodCursor and ask questions about the project was extremely satisfying, and the satisfaction stems from the idea of having someone try my project that I created from scratch and observe their reaction at the same time.
 
---
 
## A Standout Conversation
 
One of the most memorable interactions I had during my time at the Expo was with a father-son pair. The young man was a new freshman who was in the process of trying to decide between two classes he could be taking next year – CSSE (Computer Science & Software Engineering) or IED (Intro to Engineering and Design).
 
The young man had always seen these as mutually exclusive options, as one or the other choice. I had a conversation with them about both classes, highlighting the merits of each pathway and how learning programming and systems design skills in CSSE would complement the physical design, prototyping, and engineering skills that the other class would teach him.
 
By the end of our interaction, they left with an enthusiasm for taking both classes, in order to create multiple potential STEM career paths for themselves during their high school years.
 
---
 
## Takeaways
 
- Showing what your project can do through interaction is one of the most effective ways to communicate its importance; better than any slide presentation or talk on its own.
- Accessibility options often get overlooked in student projects. NodCursor proves that they are impressive and impactful while being relatively easy to implement.
- Working at our booth with Nikhil, Rohan, and Aditya made for an interesting experience. Having two strong yet different projects made for a more exciting event than either one would have on their own.
- The biggest surprise of the event was meeting the dad and son. It shows that CTE Expos aren't only a chance to display your work but also change someone's perspective on CS and engineering education.
